```python
import pandas as pd
import altair as alt

data = pd.read_csv("UR5-log-SamePayload.tsv", delimiter="\t")

data = data.drop(columns=["label1", "label2", "label3"])
data = data[1079:]

data = data.pivot(index="milestone", columns="robot_id", values="timestamp")
data["difference"] = data[0] - data[1]
data = data.reset_index()
data = data.rename(columns={0: "robot0", 1: "robot1", 2: "robot2"})

alt.Chart(data).mark_point(filled=True, size=10).encode(x="milestone", y="difference").properties(width=600, height=600)
data
```

```python
df = pd.read_csv("2022.02.02-Julian-Robot-Data-time_diff_raw-2.csv")
df["difference"] = df["robot0time"] - df["robot1time"]
df = df[df["test_run"] != 1]
df = df[df["test_run"] != 2]
df = df[df["test_run"] != 3]

alt.Chart(df).mark_point(
    filled=True,
    size=10
).encode(
    x="milestone",
    y="difference",
    color="test_run:O"
).properties(width=600, height=600)
```
